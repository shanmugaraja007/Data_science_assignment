```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("kidney_disease.csv")
data
```

```python
data.isnull().sum().sum()
```

# fill with previous value

```python
data1=data.fillna(method='pad')
data1
```

# fill with next value

```python
data2=data.fillna(method='bfill')
data2
```

# remove if any null value in row

```python
data3=data.dropna(how='any')
data3
```

# remove only if entire row having null

```python
data4=data.dropna(how='all')
data4
```

```python
data["age"].fillna(value=data["age"].mean(),inplace=True)
data
```

```python
quan=[]
for i in data.columns:
      if data[i].dtype != 'O':
            quan.append(i)
quan
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(data[quan])
df=imp.transform(data[quan])
df = pd.DataFrame(df,columns = quan)
```
